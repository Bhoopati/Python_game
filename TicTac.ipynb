```python
def cor_n(w):
    
    while w.isdigit() == False or int(w)>8: 

        w = input("please enter the valid digit")
         
        if w.isdigit == True and int(w) in range(0,9):
            
            
            break
        
    
    c = int(w)
    
    
    return c
```
